```python
import numpy as np
import control as ct
import matplotlib.pyplot as plt
```

# Primeira parte

```python
# Definindo os coeficientes do numerador e denominador da função de transferência
num = [1, 100, 2400]
den = [1, 20, 0, 0]

# Utilizando a função tf da biblioteca control para criar a função de transferência
T = ct.tf(num, den)

# Plotando o gráfico de root locus utilizando a biblioteca control
ct.root_locus_plot(T, grid = False)
plt.title('Root locus de T(s)')
plt.xlabel('Eixo real')
plt.ylabel('Eixo imaginário')
plt.show()
```

# Segunda parte

```python
num_novo = [20, 1200]
den_novo = [1, 40, 1200, 0]

T_novo = ct.tf(num_novo, den_novo)

ct.root_locus_plot(T_novo, grid = False)
plt.title('Root locus de T_novo(s)')
plt.xlabel('Eixo real')
plt.ylabel('Eixo imaginário')
plt.show()
```
