```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl 
```

Plot dos gr√°ficos

```python
# Definindo a fun√ß√£o de tranfer√™ncia G(s)
num = np.array([1, 4])
den = np.array([1, 3.2, 2.4, 0])

Valores_K = [2, 5, 9.6] # Valores de K que ir√£o ser testados

G_s = ctrl.TransferFunction(num, den) # Criando a fun√ß√£o de transfer√™ncia G(s)

for K in Valores_K:
    G_s_K = G_s * K # Multiplica√ß√£o em cascata 
    G_malha_fechada = ctrl.feedback(G_s_K, 1) # Gerando a malha fechada com realimenta√ß√£o unit√°ria
    t, y = ctrl.step_response(G_malha_fechada) # Gerando a resposta ao degrau unit√°rio 
    plt.plot(t, y, label=f'K = {K}')
    plt.title(f"Resposta ao Degrau para K = {K}")
    plt.xlabel("Tempo (s)")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.show()


```
